```python
from matplotlib import pyplot as plt

fileName = "pos.txt"
with open(fileName) as f:
    lines = f.readlines()
#
X = []
Y = []
for i in range(len(lines)):
    x, y = lines[i].split()  
    X.append(float(x))
    Y.append(float(y))
#

from matplotlib import pyplot as plt
plt.plot(X,Y, 'o')

```
